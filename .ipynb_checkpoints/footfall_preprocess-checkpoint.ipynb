# 2019-2024 Footfall

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
```

## Pre-processing

```python
file_paths = [
    "data/StationFootfall_2019.csv",
    "data/StationFootfall_2020.csv",
    "data/StationFootfall_2021.csv",
    "data/StationFootfall_2022.csv",
    "data/StationFootfall_2023_2024 .csv"
]

tfl_list = []

for path in file_paths:
    df = pd.read_csv(path, header = 0)

    df.drop(columns=['DayOfWeek'], errors = 'ignore', inplace = True)
    df['TravelDate'] = pd.to_datetime(df['TravelDate'], format = '%Y%m%d')
    df['EntryTapCount'] = pd.to_numeric(df['EntryTapCount'], errors = 'coerce')
    df['ExitTapCount'] = pd.to_numeric(df['ExitTapCount'], errors = 'coerce')

    df.dropna(subset = ['EntryTapCount', 'ExitTapCount'], inplace = True)

    tfl_list.append(df)

tfl_all = pd.concat(tfl_list, ignore_index = True)
tfl_all.reset_index(drop = True, inplace = True)
display(tfl_all.head())
```

## Aggregation and Average by Station

```python
tfl_all['TotalFlow'] = tfl_all['EntryTapCount'] + tfl_all['ExitTapCount']

total_by_station = (
    tfl_all.groupby('Station')[['EntryTapCount', 'ExitTapCount', 'TotalFlow']]
    .sum()
    .reset_index()
    .sort_values(by='TotalFlow', ascending=False)
)

mean_by_station = (
    tfl_all.groupby('Station')[['EntryTapCount', 'ExitTapCount', 'TotalFlow']]
    .mean()
    .reset_index()
    .sort_values(by='TotalFlow', ascending=False)
)

print("Top 10 busiest stations (total over 5 years):")
display(total_by_station.head(10))

print("Top 10 busiest stations (average per day):")
display(mean_by_station.head(10))

total_by_station.to_csv("StationFootfall_Total.csv", index=False)
mean_by_station.to_csv("StationFootfall_DailyAverage.csv", index=False)
```

## Visualization

```python
top_footfall = total_by_station.head(10).copy()
num_bars = len(top_footfall)
colors = cm.Set2(np.linspace(0, 1, num_bars))

plt.figure(figsize=(12, 6), dpi=100)
plt.bar(top_footfall["Station"], top_footfall["TotalFlow"], color=colors, edgecolor="white")
plt.title("Top 10 Stations by Total Footfall (20192024)", fontsize=16)
plt.xlabel("Station", fontsize=12)
plt.ylabel("Total Footfall (in Millions)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x/1e6:.0f}M'))
plt.tight_layout()
plt.show()
```
